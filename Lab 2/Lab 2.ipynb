```python
import pandas as pd
from nltk.corpus import brown
import matplotlib.pyplot as plt
```

```python
txt=open("./Corpus/515053.txt", 'r')
txtwords=txt.read().split()
```

```python
stopwords=open("./stopwords.txt", 'r').read().split()
```

```python
words=[]+txtwords
tokens={}
```

```python
for i in words:
    if i.lower() not in tokens:
        tokens[i.lower()]=1
    else:
        tokens[i.lower()]+=1
```

```python
tokens
```

```python
df=pd.DataFrame(tokens.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
# df=df.sort_values(by = 0, axis = 1)
df
df2=pd.DataFrame.from_dict(tokens, orient='index')
df2=df2.sort_values(by = 0)
df2
df.hist()
plt.savefig('save.png')
```

```python
plt.bar(tokens.keys(), tokens.values())
plt.savefig('save.png')
```

```python
my_stopwords=[]
highest=max(tokens.values())
lowest=min(tokens.values())
threshold=(highest-lowest)*0.6
threshold

for i in tokens:
    if tokens[i]>=threshold:
        my_stopwords.append(i)
match=[]
not_match=[]

for i in my_stopwords:
    if i not in stopwords:
        not_match.append(i)
    else:
        match.append(i)
```

```python
cutoff_percent=len(match)/len(stopwords)*100
cutoff_percent
```
